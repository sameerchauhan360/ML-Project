```python
# importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
heart_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/heart.csv')
```

```python
# statictical measure about the data
heart_data.describe()
```

```python
# Checking distribution of target variable
heart_data['target'].value_counts()
```

1 --> Defective

0 --> Safe

Seprating the ***feature*** & ***target***

```python
X = heart_data.drop(columns='target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
# fit the data with model
model.fit(X_train, Y_train)
```

Model Evaluation --> Accuracy Score

```python
# training data Accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training Data ', training_data_accuracy)
```

```python
# test data Accuracy
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test Data ', test_data_accuracy)
```

Building Predictive System

```python
input_data = (44,1,2,130,233,0,1,179,1,0.4,2,0,2)

# change in numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
```

```python
if (prediction[0] == 0):
  print('The Pearson does not have heart disease')
else: print('The Pearson has heart disease')
```
